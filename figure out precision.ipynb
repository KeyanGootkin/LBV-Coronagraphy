```python
import numpy as np
import corona as cr
from precision.utility import Observation
from precision.irdis import Irdis
from astropy.io import fits
import matplotlib.pyplot as plt
from glob import glob
import gootools as gt
import warnings
warnings.filterwarnings('ignore',category=RuntimeWarning)
print(cr.has_cals,np.unique(cr.mjd))
```

```python
def XMLsim(name):
    mode = 'IRDIS_CI'
    dark = 'IRD_DARK'
    flat = 'IRD_CLI_FLAT'
    sky = 'IRD_SCI_CLI_SKY'
    flux = 'IRD_FLUX_CALIB_CORO_RAW'
    observation = {'category' : 'IRD_SCI_CLI_OBJ',
                   'main' : [[mode+'_RAW',f[:-5]]    for f in cr.files if name in f and f not in cr.calibrations],
                   'cals' : [
                               {'category' : dark,
                                'main' : [[dark+'_RAW',d[:-5]]    for d in cr.darks if name in d],
                                'cals' : []},
                               {'category' : flat,
                                'main' : [[flat+'_RAW',fl[:-5]]    for fl in cr.flats if name in fl],
                                'cals' : [
                                    {'category' : dark,
                                     'main' : [[dark+'_RAW',d[:-5]]    for d in cr.darks if name in d],
                                     'cals' : []}
                                ]},
                               {'category' : sky,
                                'main' : [[sky+'_RAW',s[:-5]]    for s in cr.sky if name in s],
                                'cals' : []},
                               {'category' : flux,
                                'main' : [[flux,flx[:-5]]    for flx in cr.fluxs if name in flx],
                                'cals' : []}     
                           ]    
                  }
    return observation
test_data = [[name,XMLsim(name)]    for name in cr.targets]
```

```python
check_fits = glob("*.fits")
for [name,xml] in test_data:
    already_reduced = False
    for check in check_fits:
        if name in check:
            already_reduced = True
            print('ding!')
            
    if not already_reduced:
        try:
            i = Irdis(Observation(xml),prefix=name)
            i.reduce()
        except:
            print("whoops no darks!")
```
